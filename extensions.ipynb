#### FR SMA conditional plotting

```python
agg = read_csv('data/agg/BTC_agg_OI_FR.csv', parse_dates=['timestamp'])
agg = agg.set_index('timestamp')
agg
```

```python
scraper = CmcScraper('BTC', '24-12-2019', '31-12-2023')
```

```python
BTC_klines = scraper.get_dataframe()
BTC_klines = BTC_klines.iloc[::-1]
BTC_klines = BTC_klines.drop(['Open', 'High', 'Low', 'Volume', 'Market Cap'], axis=1)
BTC_klines = BTC_klines.reset_index(drop=True)
BTC_klines = BTC_klines.rename(columns={"Date": "timestamp"})
BTC_klines = BTC_klines.set_index('timestamp')
BTC_klines['SMA100'] = BTC_klines['Close'].rolling(100).mean()
BTC_klines = BTC_klines.dropna()
BTC_klines
```

```python
res = concat([agg, BTC_klines], axis=1)
res
```

```python
res['Color'] = ['Green' if close > sma100 else 'Red' for close, sma100 in zip(res['Close'], res['SMA100'])]

df_green = res[res['Color'] == 'Green']
df_red = res[res['Color'] == 'Red']
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_green, x=df_green.index, y='Agg Weighted FR', color='green', s=5)
sns.scatterplot(data=df_red, x=df_red.index, y='Agg Weighted FR', color='red', s=5)
plt.title('Agg Weighted FR based on Close-SMA100 condition')
plt.xlabel('Date')
plt.ylabel('Agg Weighted FR')
plt.legend(['Above SMA100', 'Below SMA100'], loc='upper left')
plt.grid(True)
plt.show()
```
